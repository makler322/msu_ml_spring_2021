```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import f1_score
# from helpers import plot_confusion_matrix
import numpy as np
import pandas as pd
from sklearn.metrics import f1_score
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.linear_model import SGDClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
train_df.head()
```

```python
train_df['k'] = ' '
train_df['u'] = train_df['url'].str.split('.')

def g(S):
    if S[0] == 'w' and S[2] == 'w':
        S = S[3:]
    return S

f = lambda x: g(' '.join(x["url"].split(".")))


train_df["disappointed"] = train_df.apply(f, axis=1)

train_df['gg'] = train_df[['disappointed', 'k', 'title', 'k', 'disappointed']].sum(axis = 1)


test_df['k'] = ' '
test_df["disappointed"] = test_df.apply(f, axis=1)

test_df['gg'] = test_df[['disappointed', 'k', 'title', 'k', 'disappointed']].sum(axis = 1)
test_df.head()
```

```python
X_train = train_df["gg"].values  # На трейн значение тайтлов

X_test = test_df["gg"].values  # На тест значение тайтлов (то что нужно)
y_train = train_df["target"].astype(int).values  # На трейн значение целей в инт
```

```python
vectorizer = CountVectorizer()
```

```python
X_train_vectorized = vectorizer.fit_transform(X_train)
```

```python
model = LogisticRegression(solver='liblinear', C = 65)
model.fit(
    X_train_vectorized,
    y_train
)

y_pred = model.predict_proba(
    X_train_vectorized
)[:, 1] >= 0.33
f1_score(y_train, y_pred)
```

```python


X_test_vectorized = vectorizer.transform(X_test)


test_df["target"] = (model.predict_proba(X_test_vectorized)[:, 1] >= 0.33).astype(bool)

test_df[["id", "target"]].to_csv("FINAL.csv", index=False)

```
